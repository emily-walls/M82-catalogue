```python
import matplotlib.pyplot as plt
import numpy as np
import pandas

#load in csv file
csvfile = pandas.read_csv('/Users/user/Documents/Catalogue/SOURCE_CAT_032026.csv')
```

Input source name | 
Please refer to the list of avalible sources to plot lightcurves for on GitHub
ie: '41.95+57.5', 'SN2008iz'

```python
source_name = '41.95+57.5'

#filter catalgoue for source name & extract values
output_source = csvfile.loc[csvfile['SOURCE_NAME'] == source_name, ['JD_OBS', 'JD_ERROR', 'TOTAL_FLUX', 'ERROR_TOTFLUX', 'FREQ_OBS', 'PEAK_FLUX', 'ERROR_PEAKFLUX',
                                                                    'PAPER', 'RESOLUTION']]

jd_values        = output_source['JD_OBS'].to_numpy()
jd_error         = output_source['JD_ERROR'].to_numpy()
total_flux_value = output_source['TOTAL_FLUX'].to_numpy()
tf_error         = output_source['ERROR_TOTFLUX'].to_numpy()
peak_flux_value  = output_source['PEAK_FLUX'].to_numpy()
pf_error         = output_source['ERROR_PEAKFLUX'].to_numpy()
frequency_obs    = output_source['FREQ_OBS'].to_numpy()
paper            = output_source['PAPER'].to_numpy()
resolution       = output_source['RESOLUTION'].to_numpy()

#group frequency values into bins
#frequency grouping 
def freq_group(freq):
    if 0.3 <= freq <= 1.0:
        return '0.4-1.0 GHz'
    elif 1.4 <= freq <= 1.7:
        return '1.4-1.7 GHz'
    elif 4.8 <= freq <= 5:
        return '4.8-5.0 GHz'
    elif 7.8 <= freq <= 9.1:
        return '7.8-9.1 GHz'
    elif freq == 15:
        return '15 GHz'
    elif freq == 22:
        return '22 GHz'
    elif freq == 43:
        return '43 GHz'
    
output_source['FREQ_BIN'] = output_source['FREQ_OBS'].apply(freq_group)


marker_for_freq = {'0.4-1.0 GHz': '<', '1.4-1.7 GHz': 'v', '4.8-5.0 GHz':'o', '7.8-9.1 GHz':'>',
                   '15 GHz': '*', '22 GHz':'8', '43 GHz':'X'}

colour_map = {'0.4-1.0 GHz': '#648FFF', '1.4-1.7 GHz': '#785EF0', '4.8-5.0 GHz':'#DC267F', '7.8-9.1 GHz':'#FE6100',
                   '15 GHz': '#FFB000', '22 GHz':'#009E73', '43 GHz':'#56B4E9'}

bin_order = ['0.4-1.0 GHz', '1.4-1.7 GHz', '4.8-5.0 GHz', '7.8-9.1 GHz', '15 GHz', '22 GHz', '43 GHz']
```

Run the next portion of code for an integrated flux plot.

```python
fig, ax = plt.subplots()

total_flux_data = output_source.dropna(subset=['TOTAL_FLUX', 'FREQ_BIN'])


for freq_bin in bin_order:
    subset = total_flux_data[total_flux_data['FREQ_BIN'] == freq_bin]

    if subset.empty:
        continue

    ax.errorbar(subset['JD_OBS'], subset['TOTAL_FLUX'], xerr=subset['JD_ERROR'], yerr=subset['ERROR_TOTFLUX'], 
                fmt=marker_for_freq.get(freq_bin, 'o'), color=colour_map.get(freq_bin, 'black'), linestyle='none', linewidth=2, capsize=6, label=freq_bin)


legend_handles = [
    plt.Line2D(
        [0], [0],
        marker    = marker_for_freq.get(freq_bin, 'o'),
        color     = colour_map.get(freq_bin, 'black'),
        linestyle = 'none',
        markersize=8,
        label=freq_bin
    )
    for freq_bin in bin_order
    if freq_bin in output_source['FREQ_BIN'].values
]

ax.legend(handles = legend_handles, title="Frequency Range")

plt.title(source_name + ' - Integrated Flux')
plt.xlabel('Date (JD)')
plt.ylabel('Integrated Flux (mJy)')
ax.set(xlim = (2441000,2458000), ylim = (0,325))
plt.show()
```

Run the next portion of code for a peak flux plot.

```python
fig, ax = plt.subplots()

peak_flux_data = output_source.dropna(subset=['PEAK_FLUX', 'FREQ_BIN'])

for freq_bin in bin_order:
    subset = peak_flux_data[peak_flux_data['FREQ_BIN'] == freq_bin]

    if subset.empty:
        continue

    ax.errorbar(subset['JD_OBS'], subset['PEAK_FLUX'], xerr=subset['JD_ERROR'], yerr=subset['ERROR_PEAKFLUX'], 
                fmt=marker_for_freq.get(freq_bin, 'o'), color=colour_map.get(freq_bin, 'black'), linestyle='none', linewidth=2, capsize=6, label=freq_bin)


legend_handles = [
    plt.Line2D(
        [0], [0],
        marker    = marker_for_freq.get(freq_bin, 'o'),
        color     = colour_map.get(freq_bin, 'black'),
        linestyle = 'none',
        markersize=8,
        label=freq_bin
    )
    for freq_bin in bin_order
    if freq_bin in output_source['FREQ_BIN'].values
]

ax.legend(handles = legend_handles, title="Frequency Range")

plt.title(source_name + ' - Peak Flux')
plt.xlabel('Date (JD)')
plt.ylabel('Peak Flux (mJy)')
ax.set(xlim = (2441000,2458000), ylim = (0,325))
plt.show()
```
